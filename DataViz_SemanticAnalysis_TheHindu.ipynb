```python
import re

def split_url(url):
    # Remove "https:", "www", and "com" using regex
    url = re.sub(r'https:|www|com|http', '', url)

    # Split the URL into words using '/'
    words = re.split(r'[!*();:@&=+$,./?-_]', url)
    words = [word for word in words if not re.search(r'\d', word)]

    # Remove empty strings from the list
    words = [word for word in words if word]

    # Remove '.' character from each word
    # words = [word.replace('.', '') for word in words]

    return words

data['Split_URL']=data['Id'].apply(split_url)
```

```python
import wordninja

def split_compound_words(word_list):
    result = [wordninja.split(word) for word in word_list]
    return result

data['words']=data['Split_URL'].apply(split_compound_words)
```

```python
def flatten_list_of_lists(lst):
    return [item for sublist in lst for item in sublist]
data['FlattenedColumn'] = data['words'].apply(flatten_list_of_lists)

# Display the updated DataFrame
# data.head()
```

```python
def filter_words(word_list):
    return [word for word in word_list if len(word) > 2]

# Apply the filtering function to the "Words" column
data['FilteredWordsSmall'] = data['FlattenedColumn'].apply(filter_words)

# Display the updated DataFrame
# data.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


english_words = set(nltk.corpus.words.words())
stop_words = set(stopwords.words('english'))
def filter_words(word_list):
    word_list = [word for word in word_list if word not in stop_words]
    return [word for word in word_list if word in english_words]

# Apply the filtering function to the "Words" column
data['FilteredWordsDictionary'] = data['FilteredWordsSmall'].apply(filter_words)

# # Display the updated DataFrame
# data.head(25)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

# Function to apply stemming to a list of words
def stem_words(word_list):
    stemmed_words = [stemmer.stem(word) for word in word_list]
    return stemmed_words

# Apply the stem_words function to the 'Words' column
data['StemmedWords'] = data['FilteredWordsDictionary'].apply(stem_words)

# Display the updated DataFrame
# data.head(25)
```

```python
data['Rank_In_degree'] = data['indegree'].rank(ascending=False)
data['Rank_Out_degree'] = data['outdegree'].rank(ascending=False)
data['Rank_closeness'] = data['closnesscentrality'].rank(ascending=False)
data['Rank_betweeness'] = data['betweenesscentrality'].rank(ascending=False)
data['Rank_pagerank'] = data['pageranks'].rank(ascending=False)
```

```python
data["Rank_sum"] = data['Rank_In_degree']+data['Rank_Out_degree']+data['Rank_closeness']+data['Rank_betweeness']+data['Rank_pagerank']
data.to_csv("Extrafeatures.csv")
```

```python
#Clustering with BERT and K-means
from sklearn.cluster import KMeans
from transformers import BertTokenizer, BertModel
import torch

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

# Function to get BERT embeddings for a list of words
def get_bert_embeddings(word_list):
    text = " ".join(word_list)
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
    return embeddings

# Apply the get_bert_embeddings function to the 'Words' column
data['BERT_Embeddings'] = data['StemmedWords'].apply(get_bert_embeddings)

# Cluster the words using K-Means clustering
n_clusters = 13  # You can adjust the number of clusters
word_embeddings = data['BERT_Embeddings'].tolist()
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
data['Cluster_BERT'] = kmeans.fit_predict(word_embeddings)

# Display the updated DataFrame
data.head(25)
```

```python
from sklearn import metrics
from sklearn.metrics import adjusted_rand_score

# Calculate Adjusted Rand Index (ARI) to measure similarity
ari = adjusted_rand_score(df['Cluster_BERT'], df['modularity_class'])
print(f"Adjusted Rand Index (ARI): {ari}")

# Create a contingency table to visualize cluster similarity
contingency_table = pd.crosstab(df['Cluster_BERT'], df['modularity_class'])

# Plot the contingency table as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap='Blues', fmt='d', cbar=True)
plt.xlabel('Modularity Class')
plt.ylabel('BERT Clusters')
plt.title('Cluster Similarity Heatmap')
plt.show()
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
nmi = normalized_mutual_info_score(df['Cluster_BERT'], df['modularity_class'])
nmi
```

```python
normalized_contingency_table = contingency_table / contingency_table.sum().sum()

# Plot the normalized contingency table as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(normalized_contingency_table, annot=True, cmap='Blues', fmt='.2f', cbar=True)
plt.xlabel('Modularity Class')
plt.ylabel('BERT Clusters')
plt.title(f'Normalized Contingency Table (NMI={nmi:.2f})')
plt.show()
```
